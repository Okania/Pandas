```python
#  Import the necessary libraries
import pandas as pd
```

```python
# Import the dataset called Euro_2012_stats_TEAM.csv and assign it to variable names euro12
euro12 = pd.read_csv('/content/Euro_2012_stats_TEAM.csv')
euro12
```

```python
# Select only the Goal column.
euro12['Goals']
```

```python
# How many team participated in the Euro2012?
euro12.shape[0]
```

```python
# What is the number of columns in the dataset?
euro12.shape[1]
```

```python
# View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline
discipline = euro12[['Team','Yellow Cards', 'Red Cards']]
discipline
```

```python
# Sort the teams by Red Cards, then to Yellow Cards
euro12.sort_values(by = ['Red Cards'])
euro12.sort_values(by = ['Yellow Cards'])
```

```python
# Calculate the mean Yellow Cards given per Team
euro12['Yellow Cards'].mean()
```

```python
#  Filter teams that scored more than 6 goals
euro12.loc[euro12.Goals > 6]
```

```python
# Select the teams that start with G
euro12.loc[euro12.Team.str.startswith('G')]['Team']
```

```python
#  Select the first 7 columns
euro12.iloc[:, :7]
```

```python
# Select all columns except the last 3
euro12.iloc[:, : -3]
```

```python
# Present only the Shooting Accuracy from England, Italy and Russia
euro12.loc[(euro12.Team == 'England') | (euro12.Team == 'Italy') | (euro12.Team == 'Russia') ][['Team','Shooting Accuracy']]
```
