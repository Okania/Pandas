```python
# Import the necessary libraries
import pandas as pd
```

```python
# This is the data given as a dictionary. Create a dataframe and assign it to a variable called army

raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 'deaths': [523, 52, 25, 616, 43, 234, 523, 62, 62, 73, 37, 35], 'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9], 'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523], 'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345], 'readiness': [1, 2, 3, 3, 2, 1, 2, 3, 2, 1, 2, 3], 'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1], 'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3], 'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}

army = pd.DataFrame.from_dict(raw_data)
army
```

```python
# Set the 'origin' colum as the index of the dataframe
army.set_index('origin', inplace = True)
```

```python
# Print only the column veterans
army.veterans
```

```python
#  Print the columns 'veterans' and 'deaths'
army[['veterans', 'deaths']]
```

```python
# Print the name of all the columns
army.columns
```

```python
#  Select the 'deaths', 'size' and 'deserters' columns from Maine and Alaska

# army.loc[['Maine','Alaska']][['deaths','size','deserters']]

army.loc[['Maine','Alaska'],['deaths', 'size' ,'deserters']]
```

```python
# Select the rows 3 to 7 and the columns 3 to 6
army.iloc[2:7, 2:6]
```

```python
# Select every row after the fourth row and all columns
army.iloc[4: , :]
```

```python
#  Select every row up to the 4th row and all columns
army.iloc[ :4, :]
```

```python
# Select the 3rd column up to the 7th column
army.iloc[: , 2:8]
```

```python
# Select rows where deaths is greater than 50
army.loc[army.deaths > 50]
```

```python
# Select rows where deaths is greater than 500 or less than 50
army.loc[(army['deaths']>500) | (army['deaths']<50) ]
```

```python
#  Select all the regiments not named "Dragoons"
army.loc[army.regiment != 'Dragoons']['regiment']
```

```python
# Select the rows called Texas and Arizona
army.loc[['Texas','Arizona']]
```

```python
# Select the third cell in the row named Arizona
army.loc['Arizona'].iloc[2]
```

```python
# Select the third cell down in the column named deaths
army.deaths[2]
```
